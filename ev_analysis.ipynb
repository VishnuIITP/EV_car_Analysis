```python
import pandas as pd

df = pd.read_csv("ev_cars.csv")

df
```

```python
df.info()
```

```python
df = df.dropna()
df.shape
```

```python
df['Price'] = df['Price'].str.replace("Rs", "")
df['Price'] = df['Price'].str.replace("Lakh", "")
df['Price'] = df['Price'].str.replace("*", "")
df['Price'] = df['Price'].str.replace("Estimated", "")
df['Price'] = df['Price'].str.strip()
df['Price']
```

```python
df['Price'] = df['Price'].str.split("-").str[0]
df['Price'] = df['Price'].astype(float)

```

```python
df.info()
df.head()
```

```python
df['Brand'] = df['Car_Name'].str.split().str[0]
df['Model'] = df['Car_Name'].str.split().str[1:].str.join("")
df.head()

```

```python
df = df[['Brand','Model','Price','Range','Power','Battery']]
df
```

```python
df['Brand'].unique()
```

```python
df['Range'] = df['Range'].str.replace("km","")
df['Range'] = df['Range'].str.strip()
df['Range'] = df['Range'].str.split("-").str[0]
df['Range'] = df['Range'].astype(float)
df.info()
```

```python
df['Range'].head()
```

```python
df['Power'] = df['Power'].str.replace("bhp",'')
df['Power'] = df['Power'].str.strip()
df['Power'] = df['Power'].str.split("-").str[0]
df['Power'] = df['Power'].astype(float)
df.info()



```

```python
df["Battery"] = df["Battery"].str.replace("kWh","")
df["Battery"] = df["Battery"].str.strip()
df["Battery"] = df["Battery"].str.split("-").str[0]
df["Battery"] = df["Battery"].astype(float)
```

```python
df.head()
```

```python
df.sort_values('Range', ascending = False).head()
```

```python
df['Price'].mean()
```

```python
df['Brand'].value_counts()
```

```python
import matplotlib.pyplot as plt

df['Brand'].value_counts().plot(kind = 'bar')

plt.xlabel("Brand")
plt.ylabel('Number of EV models')
plt.title('EV Models by Brand')

plt.show()
```

```python
plt.scatter(df['Price'],df['Range'])

plt.xlabel("Price (lakh)")
plt.ylabel("Range (km)")
plt.title("Price vs Range of Electric Vehicles")

plt.show()
```

```python
import plotly.express as px
import plotly.io as pio

pio.renderers.default = 'browser'

fig = px.scatter(df,
                 x='Price',
                 y = 'Range',
                 color = 'Brand',
                 hover_data = ['Brand', 'Model','Battery','Power'],
                 title = 'Price Vs  Range of Electric Vehicles'
                 ,height=600,
                  width=900)


fig.show()
```

```python
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "browser"

fig = px.scatter(
    df,
    x="Battery",
    y="Range",
    color="Brand",
    hover_data=["Brand","Model","Price","Power"],
    title="Battery Capacity vs Driving Range of Electric Vehicles",
    height=600,
    width=900
)

fig.show()
```

```python
import plotly.express as px
import plotly.io as pio

pio.renderers.default = 'browser'

fig = px.histogram(
    df,
    x= 'Range',
    nbins = 10,
    color = 'Brand',
    hover_data = ['Brand','Model','Price','Battery','Power'],
    title = 'Distribution of EV Driving Range in India',
    height = 600,
    width = 900)

fig.show()
```

```python
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "browser"  # chart opens in browser

# Compute average price per brand
avg_price = df.groupby("Brand")["Price"].mean().reset_index()

# Plot bar chart
fig = px.bar(
    avg_price,
    x="Brand",
    y="Price",
    color="Brand",
    hover_data=["Price"],
    title="Average EV Price by Brand (in Lakh ₹)",
    height=600,
    width=900
)

fig.show()
```
